# Extracting original waveforms from training set
This notebook extracts the origina waveforms that the training set is build from.

## Assumptions about training set
The training set is assumed to be created from timeshift interpolation. Therefore each waveform has to be compared against other time shifted waveforms. The duplicates will be removed.

Assuming the training set is in waveform form.

```python
def get_numpy_array_size(arr):
    # Get the size of each element in bytes
    element_size = arr.itemsize
    
    # Get the total number of elements in the array
    num_elements = arr.size
    
    # Calculate the total size of the array in bytes
    total_size = element_size * num_elements
    
    # Return the size in a human-readable format
    if total_size < 1024:
        return "{} bytes".format(total_size)
    elif total_size < 1048576:
        return "{:.2f} KB".format(total_size / 1024)
    elif total_size < 1073741824:
        return "{:.2f} MB".format(total_size / 1048576)
    else:
        return "{:.2f} GB".format(total_size / 1073741824)
```

```python

get_numpy_array_size(pure_train)
```

```python
num_comparisons = 2000

current_waveform = pure_train[0]

compare_array =  pure_train[:num_comparisons][:]
    
print(compare_array.shape)    

result = np.max(compare_array - current_waveform,axis=1)

print(result.shape)

plt.plot(result,".")
```
